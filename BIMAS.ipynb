```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bim = pd.read_csv("BIMAS.csv", parse_dates=["Tarih"], dayfirst=True)
df=pd.DataFrame(bim)

df.rename(columns = {'Şimdi':'Kapanış'}, inplace = True)
df.head()
```

```python
df=df.iloc[::-1]
df
```

```python
df.info()
```

```python
df["Kapanış"]=df["Kapanış"].str.replace(",",".")
df["Kapanış"]=df["Kapanış"].astype(float)
df["Yüksek"]=df["Yüksek"].str.replace(",",".")
df["Yüksek"]=df["Yüksek"].astype(float)
df["Açılış"]=df["Açılış"].str.replace(",",".")
df["Açılış"]=df["Açılış"].astype(float)
df["Düşük"]=df["Düşük"].str.replace(",",".")
df["Düşük"]=df["Düşük"].astype(float)
```

```python
df.info()
```

```python
df.set_index("Tarih")
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x=df["Tarih"],y=df["Kapanış"])
plt.plot()
```

# Yarının Kapanış Fiyatını DF'ye eklemek

```python
df["Yarın_Kapanış_Fiyat"]=df["Kapanış"].shift(-1)
df.head()
```

# 1 Gündeki Fiyat Değişikliği(Bugün-Yarın)

```python
df["Değişim"]=df["Yarın_Kapanış_Fiyat"]-df["Kapanış"]
df
```

# Günlük Getiri Kolonu Eklemek

günlük getiri=Değişim/Kapanış

```python
df["Günlük_Getiri"]=df["Değişim"]/df["Kapanış"]
df
```

# FAST AND SLOW SIGNAL

```python
df["Ma10"]=df["Kapanış"].rolling(10).mean() #fast signal
df["Ma50"]=df["Kapanış"].rolling(50).mean() #slow signal
```

# Uzun ve Yavaş Sinyali Görselleştirme

```python
plt.figure(figsize=(10,10))
sns.lineplot(x=df["Tarih"],y=df["Kapanış"],label="Kapanış")
sns.lineplot(x=df["Tarih"],y=df["Ma10"],label="Ma10")
sns.lineplot(x=df["Tarih"],y=df["Ma50"],label="Ma50")
plt.legend()
plt.show()
```

# Shares Kolonu Oluşturma(long olup olmadığını belirlemek için)

download.png

```python
df["Shares"]=[1 if df.loc[i,"Ma10"]>df.loc[i,"Ma50"] else 0 for i in df.index]
```

```python
df.tail()
```

# Günlük Kar Hesaplama
* Negatif ise para kaybederiz
* Shares == 0 ise kâr 0 a eşittir

```python
df["Günlük_Kar"]=[df.loc[i,"Yarın_Kapanış_Fiyat"]-df.loc[i,"Kapanış"] if df.loc[i,"Shares"]>0 else 0 for i in df.index]
```

```python
df.head(10)
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x=df["Tarih"],y=df["Günlük_Kar"])
```

`# 0'dan büyük alanda para kazanırız 0'dan küçük alanda para kaybederiz`

# Kümülatif Zenginliği Hesaplamak

```python
df["Zenginlik"]=df["Günlük_Kar"].cumsum()
df.tail(10)
```
